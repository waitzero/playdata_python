```python
res = []
for i in range(0, 2):
    df = pd.read_excel('a.xlsx', engine='openpyxl', sheet_name=i)
    res.append(df)

x = res[0].merge(res[1])  #국영수 쉬트와 사,지,과 쉬트 병합하여 x에 저장
t = x.sum(axis=1)         #각 행의 총합
a = x.mean(axis=1)        #각 행의 평균
x['total']=t              #총합 결과를 total컬럼으로 추가
x['avg']=a                #평균 결과를 avg 컬럼으로 추가

print(x)
wr = pd.ExcelWriter('c.xlsx', engine='xlsxwriter')
x.to_excel(wr, index=False, sheet_name='result') #sheet_name='Sheet3'
wr.save()
```

```python
#선별 진료소 엑셀 파일 내용 로드
df = pd.read_excel('선별진료소.xlsx', engine='openpyxl')
df
```

```python
#장애인 편의사항 제공 선별진료소를 시도별로 검색하는 함수
#의료기관명을 파람으로 받아 그 진료소의 시작시간, 끝시간 알려줌
def search(city):
    
    진료소 = df[df['장애인 편의사항'].isna()==False]
    return 진료소[진료소['시도']==city]
```

```python
res=search('서울')
res[['의료기관명', '주소', '평일 운영시간', '대표 전화번호']] #여러 컬럼 추출
```

```python
def ontime(name, mode):
    info = df[df['의료기관명']==name]
    return info[mode+'운영시간']
```

```python
ontime('중앙보훈병원'.'토요일')
```

```python
def 일요일운영시간(name):
    info =df[df['의료기관명']==name]
    t = info['일요일/공휴일\n운영시간']
    if t.values[0]=='미운영':
        return t.values[0]
    else:
        return ' 운영함.진료시간:'+t.values[0]
```
